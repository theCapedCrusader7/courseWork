```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# reading image
img=plt.imread('nature2.jpeg')
img.shape
```

```python
plt.imshow(img)
```

```python
R=img[:,:,0]
G=img[:,:,1]
B=img[:,:,2]
```

```python
plt.imshow(R)
```

```python
plt.imshow(G)
```

```python
plt.imshow(B)
```

```python
covR=np.cov(R,R)
covG=np.cov(G,G)
covB=np.cov(B,B)
```

```python
plt.plot(covR)
```

```python
plt.plot(covG)
```

```python
plt.plot(covB)
```

```python
# plotting the heatmap
hmR= sn.heatmap(covR)
# displaying the plotted heatmap
plt.show()
```

```python
# plotting the heatmap
hmG = sn.heatmap(covG)
# displaying the plotted heatmap
plt.show()
```

```python
# plotting the heatmap
hmB = sn.heatmap(covB)
# displaying the plotted heatmap
plt.show()
```

```python
f,axs=plt.subplots(1,3,figsize=(20,10))
axs[0].plot(covR)
axs[1].plot(covG)
axs[2].plot(covB)
```
