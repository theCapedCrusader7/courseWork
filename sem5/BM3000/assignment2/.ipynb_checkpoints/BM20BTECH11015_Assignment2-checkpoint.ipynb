### Intensity

```python
plt.hist(x=pic.ravel(), bins=256, range=[0, 255])
plt.title("Histogram Showing Pixel Intensities And Counts")
plt.ylabel("Number Of Pixels Belonging To The Pixel Intensity")
plt.xlabel("Pixel Intensity")
plt.show()
```

The above histogram shows the number of pixels have a particular intensity level. In other words we can say that this is the pdf of the intensity distribution of the picture. And since this is the plot of the intensity, we can say that this is the distribution of rods.

### Color

```python
imbgr = cv2.imread('picture.jpeg')
imrgb = cv2.cvtColor(imbgr, cv2.COLOR_BGR2RGB)
imlab = cv2.cvtColor(imbgr,cv2.COLOR_BGR2LAB)

# surface map
fig = plt.figure(3)
plt.figure(figsize=(10,20))
ax = plt.axes(projection='3d')

y = range( imlab.shape[0] )
x = range( imlab.shape[1] ) 
X, Y = np.meshgrid(x, y)

surf = ax.plot_surface( X, Y, imlab[:,:,0], cmap=cm.jet )
fig.colorbar(surf)

# Add a color bar which maps values to colors.
ax.set_xlabel("Width of the Image")
ax.set_ylabel("Height of the Image")
ax.set_zlabel("0 - 255 Color Scale")
plt.show()
```

The above plot, plots a 3D graph of the color of each pixel.

```python
# Assuming the pdfs of the rod cells to be gaussian distributions with the same variance but different means.
```

```python
data=asarray(picNew)

print(data)
```

```python
def gaussian(x, mu):
    return (np.exp(-np.power(x - mu, 2.) / (2 * np.power(30, 2.))))/(2*np.pi)

x_values = np.linspace(0, 180, 449)
for mu, sig in [(-1, 1), (0, 2), (2, 3)]:
    plt.plot(x_values, gaussian(x_values, mu))

plt.show()
```

```python
for j in range(1,180):
    for i in range(449):
        data[i]=data[i]*gaussian(i,j)
    data[j]+=data[j-1]
```
