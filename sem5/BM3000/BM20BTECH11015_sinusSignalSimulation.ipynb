```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fftfreq
from scipy.fftpack import fft, ifft
```

```python
t=np.linspace(0,2,1000)
sin1=[]
sin2=[]
sin3=[]
sin4=[]
```

```python
#  Let the first sin function be sin(5pi*t)
# Here f1 = 5 and phase is 0
def sinAngle_1(t):
    return np.sin(50*math.pi*t)

#  Let the second sin function be sin(5pi*t + pi/6)
# Here f2 = 5 and phase is 30
def sinAngle_2(t):
    return np.sin(50*math.pi*t + (math.pi/6))

#  Let the second sin function be sin(5pi*t + pi/4)
# Here f3 = 5 and phase is 45
def sinAngle_3(t):
    return np.sin(50*math.pi*t + (math.pi/4))

#  Let the second sin function be sin(5pi*t + pi/3)
# Here f2 = 5 and phase is 60
def sinAngle_4(t):
    return np.sin(50*math.pi*t + (math.pi/3))
```

```python
for i in range(len(t)):
    sin1.append(sinAngle_1(t[i]))    
    
for i in range(len(t)):
    sin2.append(sinAngle_2(t[i]))

for i in range(len(t)):
    sin3.append(sinAngle_3(t[i]))  

for i in range(len(t)):
    sin4.append(sinAngle_4(t[i]))
```

```python
plt.plot(t,sin1)
plt.title('Sin1(t)')
```

```python
plt.plot(t,sin2)
plt.title('Sin2(t)')
```

```python
plt.plot(t,sin3)
plt.title('Sin3(t)')
```

```python
plt.plot(t,sin4)
plt.title('Sin4(t)')
```

### Fourier transforms


```python
# sin1
sin1_fft = fft(sin1)
freq = fftfreq(len(sin1), d=8./1000)
# print(sig1_fft)

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.stem(freq, np.abs(sin1_fft), 'b', markerfmt=" ", basefmt="-b")
plt.title('FFT of the signal')
plt.xlim(0, 15)
plt.xlabel('Frequency (Hz)')
plt.ylabel('FFT Amplitude')
```

### Phase spectrums

after applying fourier transform on sin(wt+pfi) we get that,

for w = w0, we get, F(sin(wt+pfi)) = pi*sin(pfi) - j*pi*cos(pfi)

for w = -w0, we get, F(sin(wt+pfi)) = pi*sin(pfi) + j*pi*cos(pfi)

otherwise, 0
