```python
#calculating R(t1,t2)
def rt(image,n,x,y):
#     temp=[[0]*(n-x)]*(n-y)
    sum=0
    
    for i in range(n-y):
        for j in range(n-x):
            sum+=(image[i][j])*(image[y+i][x+j])
            
    return float(sum/float((n+x)*(n+y)))

```

```python
print(rt(temp,4,1,0))
```

```python
kfreq = np.fft.fftfreq(300) * 300
kfreq2D = np.meshgrid(kfreq, kfreq)
knrm = np.sqrt(kfreq2D[0]**2 + kfreq2D[1]**2)

knrm = knrm.flatten()
fft1Amp = fft1Amp.flatten()

kbins = np.arange(0.5, 300//2+1, 1.)
kvals = 0.5 * (kbins[1:] + kbins[:-1])
Abins, _, _ = stats.binned_statistic(knrm, fft1Amp,statistic = "mean",bins = kbins)
Abins *= np.pi * (kbins[1:]**2 - kbins[:-1]**2)

plt.plot(kvals, Abins)
plt.xlabel("$k$")
plt.ylabel("$P(k)$")
plt.tight_layout()
```
