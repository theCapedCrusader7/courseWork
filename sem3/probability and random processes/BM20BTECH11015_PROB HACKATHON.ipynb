```python
def Peaks(X):
    peaks=[]
    for i in range(1,len(X)-1):
        #if ( (X[i-1]<=X[i]) and (X[i]>=X[i+1]) ) and ( (Y[i-1]<=Y[i]) and (Y[i]>=Y[i+1]) ):
        if ( (X[i-1]<=X[i]) and (X[i]>=X[i+1]) ):
            # peaks.append((X[i],Y[i]))
            peaks.append(X[i])
    return peaks
```
