# Plotting a 4d gaussian where the 4th dimension is color

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
mu=0
sigma=1
A=np.array([[1,0,0],[0,1,0],[0,0,1]])
meanMat=np.array([mu,mu,mu])

x = np.random.randint(low=-3*sigma,high= 3*sigma, size=(200,))
y = np.random.randint(low=-3*sigma,high= 3*sigma, size=(200,))
z = np.random.randint(low=-3*sigma,high= 3*sigma, size=(200,))
```

```python
f=np.array([x,y,z])
f=np.transpose(f)

pdf=[] 

for i in range(f.shape[0]):
    pdf.append(1/np.sqrt(( (2*np.pi)**f.shape[1] )*det(A)) * np.exp(-1/2* np.transpose(f[i]-mu) @ np.linalg.inv(A) @ (f[i]-mu)))
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

color_map = cm.ScalarMappable(cmap=cm.plasma)
color_map.set_array(pdf)

img = ax.scatter(x, y, z, marker='h',s=200, c=pdf,cmap='plasma')
plt.colorbar(color_map)
ax.set_title("PDF Cloud")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
