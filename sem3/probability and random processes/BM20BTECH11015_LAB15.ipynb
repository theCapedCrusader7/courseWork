```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Exercise-A 

```python
t=[0,1,2,3,4]

x=np.random.normal(0,1,size=(len(t),10))
y=np.zeros((len(t),10))
z=np.zeros((len(t),10))

for i in range(1,len(t)):
    for j in range(x.shape[1]):
        y[i][j]=0.5*(x[i][j]+x[i-1][j])
        z[i][j]=x[i][j]-x[i-1][j]
        
crossCovarianceXY=np.zeros((5,5))
crossCovarianceYZ=np.zeros((5,5))
crossCovarianceZX=np.zeros((5,5))

for i in range(crossCovarianceXY.shape[0]):
    for j in range(crossCovarianceXY.shape[1]):
        crossCovarianceXY[i][j]=np.cov(x[j],y[i])[0][1]
        crossCovarianceYZ[i][j]=np.cov(y[j],z[i])[0][1]
        crossCovarianceZX[i][j]=np.cov(x[j],z[i])[0][1]
        
```

```python
print("The cross covariance of x and y is ","\n",crossCovarianceXY)
print("The cross covariance of y and z is ","\n",crossCovarianceYZ)
print("The cross covariance of z and x is ","\n",crossCovarianceZX)

```

```python
y1=norm.pdf(x,loc=0,scale=2)
y2=norm.pdf(y,loc=0,scale=2)
y3=norm.pdf(z,loc=0,scale=2)
```

```python
plt.scatter(x,y1)
plt.title("pdf of X")
plt.xlabel("X")
plt.ylabel("pdf")
plt.show()
```

```python
plt.scatter(y,y2)
plt.title("pdf of Y")
plt.xlabel("Y")
plt.ylabel("pdf")
plt.show()
```

```python
plt.scatter(z,y3)
plt.title("pdf of Z")
plt.xlabel("Z")
plt.ylabel("pdf")
plt.show()
```

# Exercise B

```python
tau=np.linspace(0,30,30)

p=0.4
pdf=[]

for i in tau:
    pdf.append(p*(1-p)**(i-1))
```

```python
plt.plot(tau,pdf)
plt.title("PDF")
plt.xlabel("T")
plt.ylabel("pdf")
plt.show()
```

# Exercise-C

```python
result=[]
p=0.46
for i in range(100):
    if i<100*p:
        result.append(1)
    else:
        result.append(-1)
```

```python
t=np.linspace(0,10000,10000)

total=0
xt=[]
yt=[]

for i in t:
    x=random.choice(result)
    xt.append(x)
    total+=x
    yt.append(total)
    
plt.plot(t,yt)
plt.show()
```

# Exercise-D

```python
def exponential(a,b):
    return np.exp(-1*a*b)
```

```python
x=np.arange(0,10)
y=[1/exponential(0.5,i) for i in x]
y1=[int(j) for j in y]
```

```python
time=[y1[i+1]-y1[i] for i in range(len(x)-2,0,-1)]
u=np.arange(1,9)
```

```python
plt.scatter(u,time,label='Time',color='blue') 
plt.scatter(u,55*exponential(0.5,u),label='Exponential',color='red') 
plt.legend()
plt.show()
```
