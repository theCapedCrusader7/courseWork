```python
for i in range(10):
    x=np.arange(0,i,1)
    y=expon.cdf(x, loc=0, scale=2)
    y1=1-y
plt.plot(x,y1)
plt.title('P(x>a)')
plt.show()
```

```python
for i in range(100):
    x=np.arange(0,i,1)
    y=poisson.cdf(x,mu=50,loc=0)
    y1=1-y
plt.plot(x,y1)
plt.title('P(x>a)')
plt.show()
```

```python
mu=50
for k in range(100):
    x=np.arange(0,k,1)
    x1=x*np.sqrt(mu)+mu
    x2=mu-x*np.sqrt(mu)
    y=poisson.sf(x1,mu,loc=0)
    y1=poisson.cdf(x2,mu,loc=0)
    y2=y+y1
plt.plot(x,y2)
plt.title('|P(x-E[x]|)>k*(std)')
plt.show()
```
