```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Data to PDF

```python
a=np.random.uniform(0,1,1000)
counts,bins,bars=plt.hist(a,bins=101)
plt.show()
```

```python
probability=counts/1000
pdf=probability/(bins[1]-bins[0])
x=np.linspace(0,1,101)
plt.plot(x,pdf)
plt.show()
```

# PDF to Data

```python
x=np.linspace(0,1,101)
length=x[2]-x[1]
dataFrequency=pdf*1000*length
dt=np.zeros(100)
for i in range(0,100):
    dt[i]=int(dataFrequency[i])
data=[]
i=1
print(dt)
for b in dt:
    if i<101:
        for a in range(0,int(b)):
            c=np.random.uniform(x[i-1],x[i])
            data.append(c)
    i+=1
plt.hist(data,bins=101)
plt.plot()            
```
